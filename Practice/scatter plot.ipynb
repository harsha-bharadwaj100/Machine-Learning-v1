```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("tip.csv")
print(data)
```

```python
x = data["day"]
y = data["total_bill"]
plt.scatter(x, y, c=data["size"], s=data["total_bill"], marker="o", alpha=0.5)
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Scatter Plot of Total Bill vs Days")
plt.show()
```
