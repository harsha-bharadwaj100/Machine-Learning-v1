```python
import numpy as np

# perform SVD on a random matrix
```

```python
A = np.random.rand(5, 8)
print(A)
```

```python
U, S, VT = np.linalg.svd(A)
print("U:", U, "S:", S, "VT:", VT, sep="\n")
```

```python
k = 3
u_k = U[:, :k]
s_k = np.diag(S[:k])
vt_k = VT[:k, :]
```

```python
print("u_k", u_k, "s_k", s_k, "vt_k", vt_k, sep="\n\n")
```

```python
a_reduced = np.dot(u_k, np.dot(s_k, vt_k))
print(a_reduced)
print(u_k @ s_k @ vt_k)
```

```python
import numpy as np

# Example matrix
A = np.array([[1, 2], [3, 4], [5, 6]])

# Perform SVD
U, S, Vt = np.linalg.svd(A, full_matrices=False)

# Reduced rank approximation (keeping only the first singular value)
k = 1  # Rank of the reduced approximation
Ur = U[:, :k]
Sr = np.diag(S[:k])
Vtr = Vt[:k, :]

# Reconstruct the reduced matrix
A_reduced = Ur @ Sr @ Vtr

print("Original matrix:")
print(A)
print("\nReduced matrix:")
print(A_reduced)
```
