```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv("tip.csv")
print(data)
```

```python
x = data["total_bill"]
plt.hist(x, bins=25, color="green", edgecolor="black", linestyle="dashed", alpha=0.6)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.show()
```
